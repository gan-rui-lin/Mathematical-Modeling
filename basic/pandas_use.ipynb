.loc[] 基本是基于标签来索引行列，但是也有时可以传入布尔数组

.iloc[] 基于整数值来索引行列，但是也有时可以传入布尔数组

直接使用 `__getitem__` 方法，默认获取列对象， 传入布尔数组/整数数组时，按行筛选（需与行数一致）

```python
import pandas as pd

df = pd.DataFrame(
    [[1, 2], [4, 5], [7, 8]],
    index=["cobra", "viper", "sidewinder"],
    columns=["max_speed", "shield"],
)
print("原始 df:")
print(df)

print("一些索引结果：")

print(df['shield'])

print("整数切片索引行：")
print(df[0:2])

print("返回具体值")
# 返回一个具体值
print(df.loc["cobra", "shield"])

# 返回一个 dataFrame
print(df.loc[["viper", "sidewinder"]])
print(type(df.loc[["viper", "sidewinder"]]))
# 返回一个 Series
print(df.loc["cobra":"viper", "max_speed"])
print(type(df.loc["cobra":"viper", "max_speed"]))
total = df.loc["cobra":"viper", "max_speed"].sum(axis=0)
print(total)
```

pandas 里的 `axis`


df.sum(axis=axis)  return the sum of the values over the requested axis 

axis = 0 表示沿着行的方向求和（对应的维度会消失），返回的 Series 与列名一致

```python
import numpy as np
data = np.arange(12).reshape(3, 4)

df = pd.DataFrame(data, columns=list("ABCD"))

print(df)

print("df.sum(axis=0):\n", df.sum(axis=0))

print("df.sum(axis=1):\n", df.sum(axis=1))
```

DataFrame 绘图

```python
import numpy as np

np.random.seed(123456)

ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))

ts = ts.cumsum()

ts.plot()
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))

df = df.cumsum()

plt.figure()

df.plot()
```
